```python
# import the necessary packages
from imutils import paths
import numpy as np
import argparse
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import sys
sys.argv=['']
del sys
```

```python
ap = argparse.ArgumentParser()
ap.add_argument("--i", "--/content/drive/MyDrive/IC", type=str, required=False,
	help="path to input directory of images to stitch")
ap.add_argument("-o", "-/content/drive/MyDrive/IC", type=str, required=False,
  help="path to the output image")
args = vars(ap.parse_args())
```

```python
# grab the paths to the input images and initialize our images list
img1 = cv2.imread('/content/drive/MyDrive/IC/ic1.jpg')
img2 = cv2.imread('/content/drive/MyDrive/IC/ic2.jpg')
img3=cv2.imread('/content/drive/MyDrive/IC/ic3.jpg')
img4=cv2.imread('/content/drive/MyDrive/IC/ic4.jpg')

images = []
images.append(img1)
images.append(img2)
images.append(img3)
images.append(img4)
```

```python
stitcher = cv2.createStitcher() if imutils.is_cv3() else cv2.Stitcher_create()
(status, stitched) = stitcher.stitch(images)
file = "/content/drive/MyDrive/IC/IC.jpg"
cv2.imwrite(file,stitched)
plt.imshow(stitched)
```
