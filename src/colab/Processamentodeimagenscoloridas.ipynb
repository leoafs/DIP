```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("/content/drive/MyDrive/images/baboon.png")
b1,g1,r1 = cv2.split(img1)
plt.figure(figsize=(10,10))
plt.subplot(141), plt.imshow(img1,cmap='gray'),plt.title("BGR")
plt.subplot(142), plt.imshow(b1,cmap='gray'),plt.title('B')
plt.subplot(143), plt.imshow(g1,cmap='gray'),plt.title('G')
plt.subplot(144), plt.imshow(g1,cmap='gray'),plt.title('R')
plt.show()
```

```python
img1A = cv2.cvtColor(img1,cv2.COLOR_BGR2YCrCb)
Y,Cr,Cb = cv2.split(img1A)
plt.figure(figsize=(10,10))
plt.subplot(141), plt.imshow(img1A,cmap='gray')
plt.subplot(142), plt.imshow(Y,cmap='gray'),plt.title('Y')
plt.subplot(143), plt.imshow(Cr,cmap='gray'),plt.title('Cr')
plt.subplot(144), plt.imshow(Cb,cmap='gray'),plt.title('Cb')
plt.show()
```

```python
img1B = cv2.cvtColor(img1,cv2.COLOR_BGR2HSV)
H,S,V = cv2.split(img1B)
plt.figure(figsize=(10,10))
plt.subplot(141), plt.imshow(img1B,cmap='gray')
plt.subplot(142), plt.imshow(H,cmap='gray'),plt.title('H')
plt.subplot(143), plt.imshow(S,cmap='gray'),plt.title('S')
plt.subplot(144), plt.imshow(V,cmap='gray'),plt.title('V')
plt.show()
```

```python
def bgrtocmyk(rgb):
  height, width = rgb.shape[:2]
  cmyk = np.zeros((height, width), dtype="uint16")
  r,g,b = cv2.split(rgb)
  c= 255-r
  m = 255- g
  y = 255- b
  k= np.minimum(c,m,y)
  cmyk = cv2.merge((c,m,y,k))
  return cmyk
cmyk=bgrtocmyk(img1)
c, m,y,k = cv2.split(cmyk)
plt.figure(figsize=(12,12))
plt.subplot(151), plt.imshow(cmyk,cmap='gray'),plt.title('CMYK')
plt.subplot(152), plt.imshow(c,cmap='gray'),plt.title('C')
plt.subplot(153), plt.imshow(m,cmap='gray'),plt.title('M')
plt.subplot(154), plt.imshow(y,cmap='gray'),plt.title('Y')
plt.subplot(155), plt.imshow(k,cmap='gray'),plt.title('K')
plt.show()
```
